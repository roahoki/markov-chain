```python
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
def cargar_grafo(filepath):
    """
    Carga un grafo dirigido desde un archivo de texto y lo representa como una lista de adyacencia.
    """
    grafo = defaultdict(list)
    with open(filepath, 'r') as archivo:
        for linea in archivo:
            origen, destino = map(int, linea.split())
            grafo[origen].append(destino)
    return grafo

def calcular_grados(grafo):
    """
    Calcula el grado entrante (degin) y el grado saliente (degout) de cada nodo en el grafo.
    """
    degin = defaultdict(int)
    degout = defaultdict(int)

    for origen, destinos in grafo.items():
        degout[origen] += len(destinos)
        for destino in destinos:
            degin[destino] += 1

    return degin, degout

# Cargar el grafo desde el archivo websample.txt
grafo = cargar_grafo('websample.txt')

# Calcular los grados entrante y saliente
degin, degout = calcular_grados(grafo)
```

```python
# Crear el histograma del grado entrante
def graficar_histograma_degin(degin):
    """
    Genera un histograma de la distribución del grado entrante.
    """
    # Contar la frecuencia de cada grado entrante
    frecuencias = defaultdict(int)
    for grado in degin.values():
        frecuencias[grado] += 1

    # Preparar los datos para el histograma
    x = list(frecuencias.keys())  # Valores posibles del grado entrante
    y = list(frecuencias.values())  # Frecuencia de cada grado

    # Crear el gráfico
    plt.bar(x, y, color='blue', alpha=0.7)
    plt.xlabel('Grado entrante')
    plt.ylabel('Número de sitios web')
    plt.title('Distribución del grado entrante')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Llamar a la función para graficar el histograma
graficar_histograma_degin(degin)
```
